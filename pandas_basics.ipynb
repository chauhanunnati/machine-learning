Select and Filter Data

```python
titanic_df['age']             # One column (Series)
titanic_df[['age', 'fare']]   # Multiple columns -> age, ticket fare
```

```python
titanic_df[0:5]               # Row slice (first 5 rows)
titanic_df.iloc[0]            # First row (by index)
titanic_df.loc[0]             # First row (by label)
```

loc vs iloc \
.loc[ ] → Label-based indexing \
use labels of rows and columns (like row index names, or column names) \
.iloc[ ] → Integer-location based indexing \
uses integer positions (like index 0, 1, 2…)

```python
titanic_df.loc[0]             # Gets the row with index label 0
titanic_df.loc[0, 'name']     # Gets the 'Name' column value of row with label 0
titanic_df.loc[:, ['name', 'age']]  # All rows, only 'Name' and 'Age' columns
```

```python
titanic_df.iloc[0]            # Gets the **first row** (position 0)
titanic_df.iloc[0, 1]         # Gets the value at first row, second column
titanic_df.iloc[:, [1, 2]]    # All rows, columns at position 1 and 2
```

```python
# Filter rows
titanic_df[titanic_df['sex'] == 'female']
# Filter with multiple conditions
titanic_df[(titanic_df['age'] > 30) & (titanic_df['sex'] == 'male')].head(3)
```

Grouping and Aggregation

```python
titanic_df.groupby('pclass')['survived'].sum()                # Total survivors by class
```

```python
# 38% of the passengers survived
titanic_df['survived'].mean()
```

```python
# Shows the average (mean) of numeric features like Age, etc. for each sex
titanic_df.groupby(['sex']).mean(numeric_only=True)
```

```python
titanic_df.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
titanic_df[titanic_df['age'] < 18].groupby(['sex', 'pclass']).mean(numeric_only=True)
```

Modify or Create New Columns

```python
titanic_df['age_doubled'] = titanic_df['age'] * 2 # New column
titanic_df['is_Child'] = titanic_df['age'] < 18   # Boolean column
titanic_df.head(2)
```

```python
titanic_df.drop('age_doubled', axis=1, inplace=True)  # Delete column
```

Handling Missing Data

```python
titanic_df.dropna()                            # Drop rows with any missing value
titanic_df.fillna(0)                           # Fill missing with 0
titanic_df['age'].fillna(titanic_df['age'].mean())     # Fill missing Age with mean age
```
